```python
import pandas as pd
import openpyxl as ab
```

```python
# Tes DATA
df = pd.read_csv('./dataset.csv', sep=';')
df.head()
```

```python
# Menampilkan 5 data terendah (tertinggi) dan (terendah) berdasarkan nilai akhir
df['Nilai Akhir'] = df['Nilai Formatif'] + df['Nilai UTS'] + df['Nilai UAS']
top_5 = df.sort_values(by='Nilai Akhir', ascending=False).head(5)
bottom_5 = df.sort_values(by='Nilai Akhir').head(5)

top_5, bottom_5
```

```python
# Menghitung nilai rata-rata dari setiap siswa
df['Rata-rata'] = df[['Nilai Formatif', 'Nilai UTS', 'Nilai UAS']].mean(axis=1)

df
```

```python
# Menghitung nilai rata-rata dari setiap nilai
rata_rata_formatif = df['Nilai Formatif'].mean()
rata_rata_uts = df['Nilai UTS'].mean()
rata_rata_uas = df['Nilai UAS'].mean()

print("Rata-rata Nilai Formatif:", rata_rata_formatif)
print("Rata-rata Nilai UTS:", rata_rata_uts)
print("Rata-rata Nilai UAS:", rata_rata_uas)
```

```python
# Mencari indeks siswa dengan nilai akhir tertinggi dan terendah
indeks_tertinggi = df['Nilai Akhir'].idxmax()
indeks_terendah = df['Nilai Akhir'].idxmin()

siswa_tertinggi = df.loc[indeks_tertinggi]
siswa_terendah = df.loc[indeks_terendah]

print("Siswa dengan nilai akhir tertinggi:")
print(siswa_tertinggi)

print("\nSiswa dengan nilai akhir terendah:")
print(siswa_terendah)
```

```python
# Mengidentifikasi siswa dengan nilai akhir yang sama
nilai_akhir_sama = df[df.duplicated(subset=['Nilai Akhir'], keep=False)]

nilai_akhir_sama
```

```python
# Menampilkan siswa dengan nilai akhir lebih dari 80
siswa_lebih_dari_80 = df[df['Nilai Akhir'] > 80]

siswa_lebih_dari_80
```

```python
# Mengurutkan dataframe berdasarkan "Nilai Akhir" dan "Nilai Formatif"
df.sort_values(by=['Nilai Akhir', 'Nilai Formatif'], ascending=[False, False], inplace=True)

df['Peringkat'] = range(1, len(df) + 1)
df
```

```python
# Menyimpan dataframe ke dalam file Excel
output_file = "output.xlsx"
df.to_excel(output_file, index=False)

print(f"Data telah diekspor ke {output_file}")
```
