```python
import pandas as pd
```

```python
df = pd.read_csv('./dataset.csv', sep=";")
df.head()
```

```python
df = df.iloc[:, :-1]
df['Nilai Akhir'] = df['Nilai Formatif'] * 0.3 + df['Nilai UTS'] * 0.3 + df['Nilai UAS'] * 0.4
df
```

```python
#1.Menentukan 5 peringkat siswa teratas
max_rank = df.sort_values(by=['Nilai Akhir'], ascending=False).iloc[0:5, :]
max_rank
```

```python
#2.Menentukan 5 peringkat siswa terbawah
min_rank = df.sort_values(by=['Nilai Akhir'], ascending=True).iloc[0:5,:]
min_rank
```

```python
#2.Menentukan Nilai rata-rata dari masing-masing siswa
df['Rata-Rata'] = df.iloc[:, 1:4].mean(axis=1)
df
```

```python
#3.Menetukan nilai rata-rata dari masing-masing ujian
avg = [df['Nilai Formatif'].mean(), df['Nilai UTS'].mean(), df['Nilai UAS'].mean()]
avg = pd.DataFrame(avg, index=['Nilai Formatif', 'Nilai UTS', 'Nilai UAS'], columns=['Rata-Rata'])
avg
```

```python
#4.Menampilkan Siswa yang memiliki nilai akhir tertinggi dan terendah
max_akhir = df[df["Nilai Akhir"] == df["Nilai Akhir"].max()]
min_akhir = df[df["Nilai Akhir"] == df["Nilai Akhir"].min()]

# Buat DataFrame baru
df = pd.DataFrame({
    "Siswa Nilai Tertinggi": max_akhir["Nama"],
    "Siswa Nilai Terendah": min_akhir["Nama"],
    "Nilai Terendah": min_akhir["Nilai Akhir"],
    "Nilai Tertinggi": max_akhir["Nilai Akhir"]
    
})
df = df.fillna('')
df
```

```python
#5.Menampilkan siswa yang memiliki nilai akhir yang sama
# Mengelompokkan siswa berdasarkan nilai akhir dan mendapatkan nama siswa yang memiliki nilai akhir yang sama
df['Siswa dengan Nilai Sama'] = df.groupby('Nilai Akhir')['Nama'].transform(lambda x: ', '.join(x))

# Menghilangkan baris dengan nilai akhir yang unik (tidak ada siswa lain dengan nilai yang sama)
df = df[df.duplicated(subset='Nilai Akhir', keep=False)]

# Menampilkan DataFrame
df = df[['Nilai Akhir', 'Siswa dengan Nilai Sama']].drop_duplicates()
df

```

```python
#6.Menampilkan siswa yang memiliki nilai akhir lebih dari 80
siswa_nilai_lebih_dari_80 = df[df['Nilai Akhir'] > 80]

# Membuat DataFrame baru
siswa_nilai_lebih_dari_80_df = siswa_nilai_lebih_dari_80[['Nama', 'Nilai Akhir']]

siswa_nilai_lebih_dari_80_df
```

```python
#7.Memberikan peringkat pada masing-masing siswa berdasarkan nilai akhir
df['Peringkat'] = df['Nilai Akhir'].rank(ascending=False, method='dense').astype(int)

# Membuat DataFrame
peringkat_siswa_df = df[['Nama', 'Nilai Akhir', 'Peringkat']]

peringkat_siswa_df = peringkat_siswa_df.sort_values(by='Peringkat')

peringkat_siswa_df
```
