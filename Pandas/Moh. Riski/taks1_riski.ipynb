```python
import pandas as pd
```

DATA NILAI

```python
df = pd.DataFrame({
    "Nama": ["Andi","Budi","Caca","Dedi","Euis","Fafa","Gigi","Haha","Iwan","Joko","Riyan","Alfin","Rizky","Rizal","Maulana"],
    "Nilai Formatif": [80,90,70,60,50,40,30,20,10,0,98,80,90,70,60],
    "Nilai UTS": [70,80,90,80,70,60,50,40,30,20,70,70,80,90,80],
    "Nilai UAS": [90,70,80,90,80,70,60,50,40,30,60,90,70,80,90]
})

df
```

Data Nilai Akhir

```python
df['Nilai Akhir'] = df['Nilai Formatif'] * 0.3 + df['Nilai UTS'] * 0.3 + df['Nilai UAS'] * 0.4
df
```

1. Tentukan 5 peringkat terbawah dan teratas!

```python
max_rank = df.sort_values(by=['Nilai Akhir'], ascending=True).iloc[0:5, :]
max_rank
```

```python
max_rank = df.sort_values(by=['Nilai Akhir'], ascending=False).iloc[0:5, :]
max_rank
```

2. Tentukan nilai rata-rata dari masing-masing siswa!

```python
df['Rata-Rata'] = df.iloc[:, 1:4].mean(axis=1)
df
```

3. Tentukan nilai rata-rata dari masing-masing ujian!

```python
avg = [df['Nilai Formatif'].mean(), df['Nilai UTS'].mean(), df['Nilai UAS'].mean()]
avg = pd.DataFrame(avg, index=['Nilai Formatif', 'Nilai UTS', 'Nilai UAS'], columns=['Rata-Rata'])
avg
```

4. Siapakah siswa yang memiliki nilai akhir tertinggi dan terendah.

```python
nilai_tertinggi = df["Nilai Akhir"].max()
siswa_nilai_tinggi = df[df["Nilai Akhir"] == nilai_tertinggi]
siswa_nilai_tinggi
```

```python
nilai_terendah = df["Nilai Akhir"].min()
siswa_nilai_rendah = df[df["Nilai Akhir"] == nilai_terendah]
siswa_nilai_rendah
```

5. Tampilkan siswa yang memiliki nilai akhir yang sama.

```python
nilai_sama = df[df.duplicated(subset="Nilai Akhir", keep= False)]
nilai_sama
```

6. Tampilkan siswa yang memiliki nilai akhir lebih dari 80.

```python
nilai_lebih_80 = df[df["Nilai Akhir"]>80]
nilai_lebih_80
```

7. Berikan peringkat pada masing-masing siswa berdasarkan nilai akhirnya dan tampilkan dalam bentuk dataframe!

```python
df["Peringkat"] = df["Nilai Akhir"].rank(ascending=False,method='min')
df["Peringkat"] = df["Peringkat"].astype(int)
df
```

8. export ke excel

```python
df.to_excel('Taks1_Riski.xlsx', index=False)
```
