```python
import pandas as pd
```

```python
data = pd.read_csv('./data.csv', sep=";")
data.head()
```

```python
data = data.iloc[:, :-1]
data['Nilai Akhir'] = data['Nilai Formatif'] * 0.3 + data['Nilai UTS'] * 0.3 + data['Nilai UAS'] * 0.4
data
```

```python
maxRank = data.sort_values(by=['Nilai Akhir'], ascending=False).iloc[0:5, :]
maxRank
```

```python
minRank = data.sort_values(by=['Nilai Akhir'], ascending=True).iloc[0:5,:]
minRank
```

```python
data['Rata-Rata'] = data.iloc[:, 1:4].mean(axis=1)
data
```

```python
rataRata = [data['Nilai Formatif'].mean(), data['Nilai UTS'].mean(), data['Nilai UAS'].mean()]
rataRata = pd.DataFrame(rataRata, index=['Nilai Formatif', 'Nilai UTS', 'Nilai UAS'], columns=['Rata-Rata'])
rataRata
```

```python
maxValue = data[data['Nilai Akhir'] == data['Nilai Akhir'].max()]
maxValue
```

```python
minValue = data[data['Nilai Akhir'] == data['Nilai Akhir'].min()]
minValue
```

```python
sameValue = data[data.duplicated(subset=['Nilai Akhir'])]
sameValue = data[data['Nilai Akhir'].isin(sameValue['Nilai Akhir'])]
sameValue
```

```python
value = data[data['Nilai Akhir'] > 80]
value
```

```python
data['Rank'] = data['Nilai Akhir'].rank(method='dense', ascending=False)
data = data.sort_values(by=['Rank'])
data
```
