```python
import pandas as pd
```

```python

data = pd.read_csv('./data.csv', sep=';')
data.head()
```

```python
peringkatRendah = data.sort_values(by=['Nilai Akhir']).head()
peringkatRendah
```

```python
peringkatAtas = data.sort_values(by=['Nilai Akhir'], ascending=False).head()
peringkatAtas
```

```python
data['Rata Rata'] = ((data['Nilai Formatif'] + data['Nilai UTS'] + data['Nilai UAS']) / 3).round(1)
data
```

```python
rataRata = data[['Nilai Formatif', 'Nilai UTS', 'Nilai UAS']].mean()
rataRata
```

```python
nilaiTerendah = data.sort_values(by=['Nilai Akhir']).head(1)
nilaiTerendah
```

```python
nilaiAkhirTinggi = data[data['Nilai Akhir'] > 80]
nilaiAkhirTinggi
```

```python
nilaiAkhirTertinggi = data[data['Rank'] == 1]
nilaiAkhirTertinggi
```

```python
data['Rank'] = data['Nilai Akhir'].rank(ascending=False).astype(int)
data
```
