```python
import pandas as pd 
```

```python
df = pd.DataFrame({
    "Nama": ["Andi","Budi","Caca","Dedi","Euis","Fafa","Gigi","Haha","Iwan","Joko","Riyan","Alfin","Rizky","Rizal","Maulana"],
    "Nilai Formatif": [80,90,70,60,50,40,30,20,10,0,98,80,90,70,60],
    "Nilai UTS": [70,80,90,80,70,60,50,40,30,20,70,70,80,90,80],
    "Nilai UAS": [90,70,80,90,80,70,60,50,40,30,60,90,70,80,90]
})

df
```

```python
df['Nilai Akhir'] = df['Nilai Formatif'] * 0.3 + df['Nilai UTS'] * 0.3 + df['Nilai UAS'] * 0.4
df
```

```python
peringkat_atas = df.sort_values(by=['Nilai Akhir'], ascending=False).iloc[0:5, :]
peringkat_atas
```

```python
peringkat_bawah = df.sort_values(by=['Nilai Akhir'], ascending=True).iloc[0:5,:]
peringkat_bawah
```

```python
value = df[df['Nilai Akhir'] > 80]
value
```

```python
avg = [df['Nilai Formatif'].mean(), df['Nilai UTS'].mean(), df['Nilai UAS'].mean()]
avg = pd.DataFrame(avg, index=['Nilai Formatif', 'Nilai UTS', 'Nilai UAS'], columns=['Nilai Rata-Rata'])
avg
```

```python
df['Nilai Rata-Rata'] = df.iloc[:, 1:4].mean(axis=1)
df
```
